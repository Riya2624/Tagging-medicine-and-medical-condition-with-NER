```python
import json
with open('Corona2.json', 'r') as f:
    data = json.load(f)
```

```python
training_data = []     # generating annotated dataset
for example in data['examples']:
    temp_dict = {}
    temp_dict['text'] = example['content']
    temp_dict['entities'] = []
    for annotation in example['annotations']:
        start = annotation['start']
        end = annotation['end']
        label = annotation['tag_name'].upper()
        temp_dict['entities'].append((start, end, label))
    training_data.append(temp_dict)
```

```python
from spacy.tokens import DocBin   # Create a DocBin object to hold serialized annotations
from tqdm import tqdm                   #tqdm library to create progress bars

nlp = spacy.blank("en") # # Initialize a blank English Spacy model
doc_bin = DocBin()
```

```python
from spacy.util import filter_spans   #The filter_spans function from spaCy's util module is used to filter a list of 
                                        # spans based on a condition.

# Iterate through each training example
for training_example  in tqdm(training_data): 
    text = training_example['text']
    labels = training_example['entities']
    doc = nlp.make_doc(text)        ## Create a Doc object from the raw text
    ents = []
    # Iterate through each entity in the example
    for start, end, label in labels:
        span = doc.char_span(start, end, label=label, alignment_mode="contract") #this method to contract or adjust the boundaries 
                                                                   #of the character span to fit more tightly around the annotated text
        if span is None:
            print("Skipping entity")
        else:
            ents.append(span)
    filtered_ents = filter_spans(ents) # Filter overlapping entities
    doc.ents = filtered_ents 
    doc_bin.add(doc)

doc_bin.to_disk("train.spacy") #Save the serialized DocBin to a file. Typically uses the .spacy extension
```
